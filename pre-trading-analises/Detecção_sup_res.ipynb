```python
# definindo funções 

def suporte(df1, l, n1, n2):
    for i in range(l-n1+1, l+1):
        if(df1.low[i]>df1.low[i-1]):
            return 0

    for i in range(l+1, l+n2+1):
        if (df1.low[i]<df1.low[i-1]):
            return 0
    return 1

def reistencia(df1, l, n1, n2):
    for i in range(l-n1+1,l+1):
        if(df1.high[i]<df1.high[i-1]):
            return 0 
    
    for i in range(l+1, l+n2+1):
        if (df1.high[i]>df1.high[i-1]):
            return 0

    return 1
```
